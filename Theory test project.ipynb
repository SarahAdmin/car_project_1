```python
import pandas as pd 
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('theory_test_data.csv',encoding='latin-1')
```

```python
grade_data = df['Status'].value_counts()
```

```python
grade_data.plot(kind='pie',legend=True,autopct='%1.1f%%',title='Theory Test Results')
```

```python
grade_data.plot(kind='bar',color='rebeccapurple',ylabel='Frequency',xlabel='Grade',title='Theory Test Results')
```

Most of the students has passed their driving theory test and some of the students failed their test.
