```python
import pandas as pd 
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('car_data.csv',encoding='latin-1')
```

```python
make_data = df['Make'].value_counts()
```

```python
make_data.plot(kind='pie',autopct='%1.1f%%',title='Car Brands',legend=False)
```

The most popular car brand is Ford because the make is a biggest car brand to audiences. The second biggest car brand is Toyota and the third biggest car brand is Volkswagen. Other car brands has the least popular frequencies than biggest car brands.

```python
model_data = df['Model'].value_counts()
```

```python
model_data.plot(kind='pie',autopct='%1.1f%%',title='Car Models',legend=False)
```

The biggest car model is Fiesta because this car model from Ford is a most popular model to audiences. The second biggest car model is Yaris and the third biggest car model is T Cross. Other car models has the same frequencies.

```python
color_data = df['Colour'].value_counts()
```

```python
color_data.plot(kind='pie',autopct='%1.1f%%',title='Car Colours',legend=False)
```

The most popular colour is White and these other car colours has the same frequencies. 

```python
fuel_data = df['Fuel Type'].value_counts()
```

```python
fuel_data.plot(kind='pie',autopct='%1.1f%%',title='Car Fuel Types',legend=True)
```

All the cars are Unleaded Fuel type than Diesel and Electric.

```python
transmission_data = df['Transmission'].value_counts()
```

```python
transmission_data.plot(kind='pie',autopct='%1.1f%%',title='Car Transmissions',legend=True)
```

The biggest transmission is Manual because most of the cars are manual for the audiences and the second biggest is Automatic because audiences interested in automatic cars than manual.
