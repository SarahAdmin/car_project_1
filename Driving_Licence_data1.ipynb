```python
import pandas as pd
df = pd.read_csv('disabled-driving-licence-data.csv',encoding='latin-1')
df['Year'] = df['Year'].astype(str)
df['Full car driving licence'] = df['Full car driving licence'].str.replace(',', '').astype(int)
df['All people'] = df['All people'].str.replace(',','').astype(int)
df['Unweighted sample size: individuals [notes 4, 5]'] = df['Unweighted sample size: individuals [notes 4, 5]'].str.replace(',','').astype(int)
```

```python
df.style.background_gradient(cmap='PuBu')
```

```python
nonDis = df.where(df['Disability status'] == 'Not disabled').dropna()
```

```python
nonDis.style.background_gradient(cmap='RdPu')
```

```python
disDF = df.where(df['Disability status'] == 'Disabled').dropna()
```

```python
disDF.style.background_gradient(cmap='RdPu')
```
